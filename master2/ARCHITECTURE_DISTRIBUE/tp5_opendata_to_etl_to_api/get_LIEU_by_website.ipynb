```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from bs4 import BeautifulSoup as BS
import html
import json

url_lieu = "https://festival-avignon.com/fr/edition-2025/programmation/par-lieu"
response = requests.get(url_lieu)
soup = BeautifulSoup(response.text, "html.parser")
tag_places = soup.find("program-by-place")
raw_places = tag_places.get(":places")
decoded_places = html.unescape(raw_places)
places = json.loads(decoded_places)

data = []
for place in places:
    print({
        "id": place["id"],
        "title": place["title"],
        "slug": place["titleSlug"],
        "address": place["location"]["address"],
        "latitude": place["location"]["latitude"],
        "longitude": place["location"]["longitude"],
    })
    data.append({
        "id": place["id"],
        "title": place["title"],
        "slug": place["titleSlug"],
        "address": place["location"]["address"],
        "latitude": place["location"]["latitude"],
        "longitude": place["location"]["longitude"],
    })

df_places = pd.DataFrame(data)
print(df_places)
df_places.to_csv("lieu_lat_long.csv", index=False, encoding="utf-8")




# url = "https://festival-avignon.com/fr/edition-2025/programmation/par-categorie?cat=1001"

# response = requests.get(url)

# soup = BeautifulSoup(response.text, "html.parser")
# tag_categorie = soup.find("program-by-category")

# # 2. Nettoyage des champs HTML (summary, description, etc.)
# def clean_html(html_text):
#     if not html_text:
#         return ""
#     soup = BS(html_text, "html.parser")
#     return soup.get_text(separator=" ", strip=True)

# ############# GET EVENTS
# events_raw = tag_categorie.get(":events")
# events_json = html.unescape(events_raw)
# events = json.loads(events_json)
# # 3. Extraction propre events
# cleaned_events = []
# for event in events:
#     cleaned_event = {
#         "id": event.get("id"),
#         "title": event.get("title"),
#         "subtitle": html.unescape(event.get("subtitle", "")),
#         "summary": clean_html(event.get("summary", "")),
#         "description": clean_html(event.get("description", ""))
#     }
#     cleaned_events.append(cleaned_event)

# # 4. Affichage des premiers r√©sultats
# for ev in cleaned_events[:3]:  # affiche les 3 premiers
#     print(f"üîπ {ev['title']}")
#     print(f"   Sous-titre : {ev['subtitle']}")
#     print(f"   R√©sum√©     : {ev['summary']}")
#     print(f"   Description: {ev['description'][:200]}...")  # tronqu√©
#     print("‚Äî" * 50)

#############" PLACES"
```
