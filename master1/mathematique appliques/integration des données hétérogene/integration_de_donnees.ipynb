```python
# EXERCICE 1:
import itertools
from scipy.spatial import distance
import pandas as pd
import numpy as np

dataframes = ["./A.csv", "./B.csv", "./C.csv"]

# distance euclidienne
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b)**2))

csv_A = pd.read_csv(dataframes[0], index_col = 0).select_dtypes(include='float64')
csv_B = pd.read_csv(dataframes[1], index_col = 0).select_dtypes(include='float64')
csv_C = pd.read_csv(dataframes[2], index_col = 0).select_dtypes(include='float64')

col_csvA = csv_A.columns
col_csvB = csv_B.columns
col_csvC = csv_C.columns

paires_col_euclidien_A_B = []
paires_col_chebyshev_A_B = []
paires_col_scalaire_A_B = []
for i in itertools.product(col_csvA, col_csvB, repeat = 1):
    col1 = csv_A[i[0]]
    col2 = csv_B[i[1]]
    col_standard1 = (col1 - col1.mean()) / col1.std()
    col_standard2 = (col2 - col2.mean()) / col2.std()
    euclidien = euclidean_distance(col_standard1, col_standard2)
    distance_chebyshev = distance.chebyshev(col_standard1, col_standard2)
    scalaire = np.dot(col_standard1,col_standard2)
    if euclidien < 10: 
            paires_col_euclidien_A_B.append((i, euclidien))
    if distance_chebyshev < 1:
        paires_col_chebyshev_A_B.append((i, distance_chebyshev))
    if scalaire < 7 and scalaire > -7:
        paires_col_scalaire_A_B.append((i, scalaire))

paires_col_euclidien_A_C = []
paires_col_chebyshev_A_C = []
paires_col_scalaire_A_C = []
for i in itertools.product(col_csvA, col_csvC, repeat = 1):
    col1 = csv_A[i[0]]
    col2 = csv_C[i[1]]
    col_standard1 = (col1 - col1.mean()) / col1.std()
    col_standard2 = (col2 - col2.mean()) / col2.std()
    euclidien = euclidean_distance(col_standard1, col_standard2)
    distance_chebyshev = distance.chebyshev(col_standard1, col_standard2)
    scalaire = np.dot(col_standard1,col_standard2)
    if euclidien < 10: 
            paires_col_euclidien_A_C.append((i, euclidien))
    if distance_chebyshev < 1:
        paires_col_chebyshev_A_C.append((i, distance_chebyshev))
    if scalaire < 7 and scalaire > -7:
        paires_col_scalaire_A_C.append((i, scalaire))

paires_col_euclidien_B_C = []
paires_col_chebyshev_B_C = []
paires_col_scalaire_B_C = []
for i in itertools.product(col_csvB, col_csvC, repeat = 1):
    col1 = csv_B[i[0]]
    col2 = csv_C[i[1]]
    col_standard1 = (col1 - col1.mean()) / col1.std()
    col_standard2 = (col2 - col2.mean()) / col2.std()
    euclidien = euclidean_distance(col_standard1, col_standard2)
    distance_chebyshev = distance.chebyshev(col_standard1, col_standard2)
    scalaire = np.dot(col_standard1,col_standard2)
    if euclidien < 10: 
            paires_col_euclidien_B_C.append((i, euclidien))
    if distance_chebyshev < 1:
        paires_col_chebyshev_B_C.append((i, distance_chebyshev))
    if scalaire < 7 and scalaire > -7:
        paires_col_scalaire_B_C.append((i, scalaire))

print("paires_col_euclidien_A_B")
print(paires_col_euclidien_A_B)
print("paires_col_euclidien_A_C")
print(paires_col_euclidien_A_C)
print("paires_col_euclidien_B_C")
print(paires_col_euclidien_B_C)
print("///")
print("paires_col_chebyshev_A_B")
print(paires_col_chebyshev_A_B)
print("paires_col_chebyshev_A_C")
print(paires_col_chebyshev_A_C)
print("paires_col_chebyshev_B_C")
print(paires_col_chebyshev_B_C)
print("///")
print("paires_col_scalaire_A_B")
print(paires_col_scalaire_A_B)
print("paires_col_scalaire_A_C")
print(paires_col_scalaire_A_C)
print("paires_col_scalaire_B_C")
print(paires_col_scalaire_B_C)
```

```python
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b)**2))

def similarite_structurelle_contenu(csv1, csv2):
    #return np.sqrt(np.sum((a - b)**2))
    col_csv1 = csv1.columns
    col_csv2 = csv2.columns

    paires_similarite_structurelle_jarowinkler = []
    paires_similarite_structurelle_levenshtein = []
    paires_similarite_contenu_jarowinkler = []
    paires_similarite_contenu_levenshtein = []
    paires_similarite_contenu_euclidien = []
    paires_similarite_contenu_chebyshev = []
    paires_similarite_contenu_scalaire = []
    paires_sans_similarite = []
    for i in itertools.product(col_csv1, col_csv2, repeat = 1):
        col1 = csv1[i[0]]
        col2 = csv2[i[1]]
        col1_type = col1.dtypes
        col2_type = col2.dtypes

        if col1_type != col2_type or col1_type == "int64" or col2_type == "int64":
            #print(col1_type, "!=", col2_type, "OR", col1_type, "== int64 OR",  col2_type, "== int64")
            continue

        similarite_structurelle_jarowinkler = jaro_winkler_similarity(i[0], i[1])
        if similarite_structurelle_jarowinkler > 0.7:
            paires_similarite_structurelle_jarowinkler.append((i, similarite_structurelle_jarowinkler))
            #print(similarite_structurelle_jarowinkler, "structurel jarowinkler", i)
            continue

        similarite_structurelle_levenshtein = calcul_levenshtein(i[0], i[1])
        if similarite_structurelle_levenshtein < 3:
            paires_similarite_structurelle_levenshtein.append((i, similarite_structurelle_levenshtein))
            #print(similarite_structurelle_levenshtein, "structurel levenshtein", i)
            continue
            
        #print("sans similaritÃ© structurelle", i)
        
        if col1_type == "object" and col2_type == "object":
            #print("all object")
            pyjarowinkler_results = 0
            for index in range(3000):
                pyjarowinkler_results += jaro_winkler_similarity(col1[index], col2[index])
                #print(pyjarowinkler_results, col1[index],  col2[index])
        
            pyjarowinkler_result = pyjarowinkler_results/3000
            #print(pyjarowinkler_result)
            if pyjarowinkler_result > 0.6:
                paires_similarite_contenu_jarowinkler.append((i, pyjarowinkler_result))
                continue
                
            levenshtein_results = 0
            for index in range(3000):
                levenshtein_results += calcul_levenshtein(col1[index], col2[index])
                #print(levenshtein_results, col1[index],  col2[index])

            levenshtein_result = levenshtein_results/3000
            #print(levenshtein_result)
            if levenshtein_result < 3:
                paires_similarite_contenu_levenshtein.append((i, levenshtein_result))
                continue
                
            
        if col1_type == "float64" and col2_type == "float64":
            #print("all float64")
            col_standard1 = (col1 - col1.mean()) / col1.std()
            col_standard2 = (col2 - col2.mean()) / col2.std()
            
            euclidien = euclidean_distance(col_standard1, col_standard2)
            if euclidien < 10: 
                paires_similarite_contenu_euclidien.append((i, euclidien))
                continue   
                    
            distance_chebyshev = distance.chebyshev(col_standard1, col_standard2)
            if distance_chebyshev < 1:
                paires_similarite_contenu_chebyshev.append((i, distance_chebyshev))
                continue 

            scalaire = np.dot(col_standard1,col_standard2)
            if scalaire < 7 and scalaire > -7:
                paires_similarite_contenu_scalaire.append((i, scalaire))  
                continue 
            
        paires_sans_similarite.append(i)
            
    result = {}
    result["paires_similarite_structurelle_jarowinkler"] = paires_similarite_structurelle_jarowinkler
    result["paires_similarite_structurelle_levenshtein"] = paires_similarite_structurelle_levenshtein
    result["paires_similarite_contenu_jarowinkler"] = paires_similarite_contenu_jarowinkler
    result["paires_similarite_contenu_levenshtein"] = paires_similarite_contenu_levenshtein
    result["paires_similarite_contenu_euclidien"] = paires_similarite_contenu_euclidien
    result["paires_similarite_contenu_chebyshev"] = paires_similarite_contenu_chebyshev
    result["paires_similarite_contenu_scalaire"] =  paires_similarite_contenu_scalaire
    result["paires_sans_similarite"] = paires_sans_similarite
    return result

print("--------------------------A B------------------------")
result_similarite_AB = similarite_structurelle_contenu(exo5_csv_A, exo5_csv_B)
print("//// structurelle_jarowinkler ////")
print(result_similarite_AB["paires_similarite_structurelle_jarowinkler"])
print("//// structurelle_levenshtein ////")
print(result_similarite_AB["paires_similarite_structurelle_levenshtein"])
print("//// contenu_jarowinkler ////")
print(result_similarite_AB["paires_similarite_contenu_jarowinkler"])
print("//// contenu_levenshtein ////")
print(result_similarite_AB["paires_similarite_contenu_levenshtein"])
print("//// contenu_euclidien ////")
print(result_similarite_AB["paires_similarite_contenu_euclidien"])
print("//// contenu_chebyshev ////")
print(result_similarite_AB["paires_similarite_contenu_chebyshev"])
print("//// contenu_scalaire ////")
print(result_similarite_AB["paires_similarite_contenu_scalaire"])
print("//// sans_similarite ////")
print(result_similarite_AB["paires_sans_similarite"])

print("--------------------------A C--------------------------")
result_similarite_AC = similarite_structurelle_contenu(exo5_csv_A, exo5_csv_C)
print("//// structurelle_jarowinkler ////")
print(result_similarite_AC["paires_similarite_structurelle_jarowinkler"])
print("//// structurelle_levenshtein ////")
print(result_similarite_AC["paires_similarite_structurelle_levenshtein"])
print("//// contenu_jarowinkler ////")
print(result_similarite_AC["paires_similarite_contenu_jarowinkler"])
print("//// contenu_levenshtein ////")
print(result_similarite_AC["paires_similarite_contenu_levenshtein"])
print("//// contenu_euclidien ////")
print(result_similarite_AC["paires_similarite_contenu_euclidien"])
print("//// contenu_chebyshev ////")
print(result_similarite_AC["paires_similarite_contenu_chebyshev"])
print("//// contenu_scalaire ////")
print(result_similarite_AC["paires_similarite_contenu_scalaire"])
print("//// sans_similarite ////")
print(result_similarite_AC["paires_sans_similarite"])

print("---------------------------B C-------------------------")
result_similarite_BC = similarite_structurelle_contenu(exo5_csv_B, exo5_csv_C)
print("//// structurelle_jarowinkler ////")
print(result_similarite_BC["paires_similarite_structurelle_jarowinkler"])
print("//// structurelle_levenshtein ////")
print(result_similarite_BC["paires_similarite_structurelle_levenshtein"])
print("//// contenu_jarowinkler ////")
print(result_similarite_BC["paires_similarite_contenu_jarowinkler"])
print("//// contenu_levenshtein ////")
print(result_similarite_BC["paires_similarite_contenu_levenshtein"])
print("//// contenu_euclidien ////")
print(result_similarite_BC["paires_similarite_contenu_euclidien"])
print("//// contenu_chebyshev ////")
print(result_similarite_BC["paires_similarite_contenu_chebyshev"])
print("//// contenu_scalaire ////")
print(result_similarite_BC["paires_similarite_contenu_scalaire"])
print("//// sans_similarite ////")
print(result_similarite_BC["paires_sans_similarite"])
```
