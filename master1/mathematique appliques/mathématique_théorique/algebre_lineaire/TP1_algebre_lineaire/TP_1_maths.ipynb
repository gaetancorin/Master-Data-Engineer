```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#exercice 1
plt.axis('equal')
plt.quiver(0,0,2,1, angles = 'xy', scale_units = 'xy', scale = 1)
plt.quiver(0,0,3,4, angles = 'xy', scale_units = 'xy', scale = 1)
plt.show()
```

```python
#exercice 2 et 3
def viz2d(vectors, labels):
    plt.axis('equal')
    for vector in vectors:
        plt.quiver(0,0,vector[0],vector[1], angles = 'xy', scale_units = 'xy', scale = 1, color= np.random.rand(3,))
    plt.legend(labels)
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    plt.show()

vectors = [[np.random.randint(0, 10), np.random.randint(0, 10)], [np.random.randint(0, 10), np.random.randint(0, 10)]]
labels = ['Vecteur U', 'Vecteur V']
viz2d(vectors, labels)
```

```python
#exercice 4

def viz2d(vectorU, vectorV, labels):
    plt.axis('equal')
    plt.quiver(0,0,vectorU[0],vectorU[1], angles = 'xy', scale_units = 'xy', scale = 1, color= np.random.rand(3,))
    plt.quiver(0,0,vectorV[0],vectorV[1], angles = 'xy', scale_units = 'xy', scale = 1, color= np.random.rand(3,))
    #exo4.a
    #plt.quiver(0,0,vectorU[0] + 0,vectorU[1], angles = 'xy', scale_units = 'xy', scale = 1, color= np.random.rand(3,))
    #exo4.b
    plt.quiver(0,0,vectorU[0]+(-vectorV[0]), vectorU[1]+(-vectorV[1]), angles = 'xy', scale_units = 'xy', scale = 1, color= np.random.rand(3,))
    #exo4.c
    plt.quiver(0,0,vectorV[0] - vectorU[0],vectorV[1]- vectorU[1], angles = 'xy', scale_units = 'xy', scale = 1, color= np.random.rand(3,))
    #exo4.d
    plt.quiver(0,0,vectorV[0] + (-vectorU[0]),vectorV[1] + (-vectorU[1]), angles = 'xy', scale_units = 'xy', scale = 1, color= np.random.rand(3,))
    #exo4.e
    plt.quiver(0,0,vectorU[0] + vectorV[0],vectorU[1] + vectorV[0], angles = 'xy', scale_units = 'xy', scale = 1, color= np.random.rand(3,))
    plt.legend(labels)
    plt.xlim(-10,15)
    plt.ylim(-10,15)
    plt.show()

vectorU = [np.random.randint(0, 10), np.random.randint(0, 10)]
vectorV = [np.random.randint(0, 10), np.random.randint(0, 10)]
labels = ['Vecteur U', 'Vecteur V', "par0", "u+(-v)"]
labels = ['Vecteur U ' +str(vectorU[0])+","+str(vectorU[1]),
          'Vecteur V '+str(vectorV[0])+","+str(vectorV[1]),
          "u+(-v) = "+str(vectorV[0]), "v-u", "v+(-u)", "u+v"]
viz2d(vectorU, vectorV, labels)
```
