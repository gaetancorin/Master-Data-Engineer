```python
import pandas as pd
import os
path = r"C:\Users\gaeta\Documents\Ynov\open_data\challenge\2_df2_calcul_moyenne_impot\input"
file = "immobilier-haute-garonne-31.xls"
df = pd.read_excel(os.path.join(path, file), sheet_name='Données', header=6)
df.head()
# df.dtypes

df = df.rename(columns={'Unnamed: 0': 'Date',
                        'Régime de droit commun (mutations à titre onéreux)': 'Regime_droit_commun',
                        'Régime dérogatoire (mutations à titre onéreux ou gratuit)': 'Regime_derogatoire',
                        'Unnamed: 3': 'Inscriptions_hypotheques'})
df = df[['Date', 'Regime_droit_commun', 'Regime_derogatoire', 'Inscriptions_hypotheques']]

df = df.sort_values(by='Date', ascending=True).reset_index(drop=True)
df.head(20)
```

```python
df['pourcentage_droit_commun'] = (df['Regime_droit_commun'] / (df['Regime_droit_commun'] + df['Regime_derogatoire'] + df['Inscriptions_hypotheques']))*100
df['pourcentage_regime_derogatoire'] = (df['Regime_derogatoire'] / (df['Regime_droit_commun'] + df['Regime_derogatoire'] + df['Inscriptions_hypotheques']))*100
df['pourcentage_inscriptions_hypotheques'] = (df['Inscriptions_hypotheques'] / (df['Regime_droit_commun'] + df['Regime_derogatoire'] + df['Inscriptions_hypotheques']))*100

df.head(20)
```

```python
chemin_fichier_csv = r'C:\Users\gaeta\Documents\Ynov\open_data\challenge\2_df2_calcul_moyenne_impot\outpout\all_dates_pourcentage_impots_par_categorie.csv'

df.to_csv(chemin_fichier_csv, index=False)
```
