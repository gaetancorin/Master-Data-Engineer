```python
import pandas as pd

chemin_fichier_2019 = r"C:\Users\gaeta\Documents\Ynov\master1\open_data\3_df1_df2_impot_par_vente\input\df_agg2019.csv"
chemin_fichier_2020 = r"C:\Users\gaeta\Documents\Ynov\master1\open_data\3_df1_df2_impot_par_vente\input\df_agg2020.csv"
chemin_fichier_2021 = r"C:\Users\gaeta\Documents\Ynov\master1\open_data\3_df1_df2_impot_par_vente\input\df_agg2021.csv"
chemin_fichier_2022 = r"C:\Users\gaeta\Documents\Ynov\master1\open_data\3_df1_df2_impot_par_vente\input\df_agg2022.csv"
chemin_fichier_2023 = r"C:\Users\gaeta\Documents\Ynov\master1\open_data\3_df1_df2_impot_par_vente\input\df_agg2023.csv"

df_agg2019 = pd.read_csv(chemin_fichier_2019)
df_agg2020 = pd.read_csv(chemin_fichier_2020)
df_agg2021 = pd.read_csv(chemin_fichier_2021)
df_agg2022 = pd.read_csv(chemin_fichier_2022)
df_agg2023 = pd.read_csv(chemin_fichier_2023)

df_agg = pd.concat([df_agg2019, df_agg2020, df_agg2021, df_agg2022, df_agg2023], ignore_index=True)
df_agg = df_agg.drop(columns=['index', "Index"])
df_agg.head(3)
```

```python
df_agg['date_mutation'] = pd.to_datetime(df_agg['date_mutation'])
df_agg['mois_annee'] = df_agg['date_mutation'].dt.to_period('M')
df_agg.head(3)
```

```python
chemin_fichier = r"C:\Users\gaeta\Documents\Ynov\master1\open_data\3_df1_df2_impot_par_vente\input\all_dates_pourcentage_impots_par_categorie.csv"
df_impot = pd.read_csv(chemin_fichier)

df_impot['Date'] = pd.to_datetime(df_impot['Date'])
df_impot['mois_annee'] = df_impot['Date'].dt.to_period('M')
df_impot.head(3)
```

```python
df_result = pd.merge(df_agg, df_impot, on='mois_annee', how='inner')
df_result.head(3)
```

```python
df_result['PE'] = df_result['valeur_fonciere']*(df_result['pourcentage_droit_commun']/100)*(4.50/100)+df_result['valeur_fonciere']*(df_result['pourcentage_regime_derogatoire']/100)*(0.66/100) +df_result['valeur_fonciere']*(df_result['pourcentage_inscriptions_hypotheques']/100)*(0.7/100)
df_result['PE'] = df_result['PE'].round().astype('int64')

df_result = df_result.drop(columns=['mois_annee', 'Date', 'Regime_droit_commun', 'Regime_derogatoire', 'Inscriptions_hypotheques','pourcentage_droit_commun', 'pourcentage_regime_derogatoire', 'pourcentage_inscriptions_hypotheques'])
df_result = df_result.sort_values(by='date_mutation', ascending=True).reset_index(drop=True)
df_result.head(15)
```

```python
path = r"C:\Users\gaeta\Documents\Ynov\master1\open_data\3_df1_df2_impot_par_vente\outpout\df_all_years_with_PE.csv"
df_result.to_csv(path, index=False)
```
