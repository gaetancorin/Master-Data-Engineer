```python
import pandas as pd
import numpy as np
import dask.dataframe as dd
import os
```

```python
# Charger le fichier CSV en utilisant Dask
df1 = dd.read_csv(r"C:\Users\gaeta\Documents\YNOV\master1\open_data\visualisation\df_ti_pe_complet.csv")

# Coordonnées de Toulouse
toulouse_lon = 1.431673
toulouse_lat = 43.596381

# Fonction pour calculer la distance entre deux points géographiques en utilisant la formule de Haversine
def haversine(lon1, lat1, lon2, lat2):
    R = 6371000  # Rayon de la Terre en mètres
    phi1 = np.radians(lat1)
    phi2 = np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)

    a = np.sin(delta_phi / 2.0) ** 2 + np.cos(phi1) * np.cos(phi2) * np.sin(delta_lambda / 2.0) ** 2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    meters = R * c  # Résultat en mètres
    return meters

# Appliquer la fonction de calcul de distance de manière vectorisée avec Dask
df1['distance à Toulouse'] = df1.map_partitions(lambda df: haversine(
    toulouse_lon, toulouse_lat, df['longitude'], df['latitude']
), meta=('float64'))
```

```python
# Calculer le résultat
df1 = df1.compute()
```
