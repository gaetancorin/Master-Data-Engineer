### Fonctions

```python
def resize_and_pad_image(image, target_size):

    # Get the original image dimensions
    original_height, original_width = image.shape[:2]

    # Calculate the ratio of the target size to the original size
    ratio = min(target_size[0] / original_width, target_size[1] / original_height)

    # Calculate the new dimensions for resizing
    new_width = int(original_width * ratio)
    new_height = int(original_height * ratio)

    # Resize the image
    resized_image = cv2.resize(image, (new_width, new_height))

    # Create a white canvas of the target size
    canvas = np.ones((target_size[1], target_size[0], 3), dtype=np.uint8) * 255

    # Calculate the position to place the resized image in the canvas
    x_offset = (target_size[0] - new_width) // 2
    y_offset = (target_size[1] - new_height) // 2

    # Paste the resized image onto the canvas
    canvas[y_offset:y_offset+new_height, x_offset:x_offset+new_width] = resized_image

    # Write the final image
    return canvas
```

```python
def is_div_20(num):
    if num % 20 == 0:
        return True
    else:
        return False
```

```python
def extract_frames_29fps(input_video, video_name, output_folder):
    cam = cv2.VideoCapture(input_video) # Ouvrir la vidéo avec OpenCV
    fps = int(cam.get(cv2.CAP_PROP_FPS)) # Récupérer la fréquence d'images par seconde (FPS)
    print(fps)
    frame_count = 0 # Initialiser la variable de compteur d'images
    while(True):
        ret, frame = cam.read() # Lire chaque frame de la vidéo
        if not ret: # Vérifier si la lecture de la vidéo est terminée
            break
        frame_count += 1 # Incrémenter le compteur de frames
        if frame_count % fps == 20:
            frame_filename = f"{output_folder}/{video_name}_000{frame_count // fps}.png"
            cv2.imwrite(frame_filename, frame)
            print(f"Image enregistrée : {frame_filename}")
    cam.release() # Libérer la vidéo
    cv2.destroyAllWindows()
```

```python
def extract_frames_60fps_resize_360p(input_video, video_name, output_folder):
    cam = cv2.VideoCapture(input_video) # Ouvrir la vidéo avec OpenCV
    fps = int(cam.get(cv2.CAP_PROP_FPS)) # Récupérer la fréquence d'images par seconde (FPS)
    print(fps)
    target_size = (852, 852)
    frame_count = 0 # Initialiser la variable de compteur d'images
    while(True):
        ret, frame = cam.read() # Lire chaque frame de la vidéo
        if not ret: # Vérifier si la lecture de la vidéo est terminée
            break
        frame_count += 1 # Incrémenter le compteur de frames
        
        if frame_count % 20 == 0: # Je veux 3 images par seconde, ainsi, je veux récupérer les images dont les frames sont un multiple de 20
            frame_filename = f"{output_folder}/{video_name}_000{int(frame_count/20)}.png"
            frame = resize_and_pad_image(frame, target_size)
            cv2.imwrite(frame_filename, frame)
            print(f"Image enregistrée : {frame_filename}")
    cam.release() # Libérer la vidéo
    cv2.destroyAllWindows()
```

## Fonctions de déformation de l'image

```python
# Fonction pour ajuster la luminosité
def adjust_brightness(image, factor: float):
    enhancer = ImageEnhance.Brightness(image)
    return enhancer.enhance(factor)

# Fonction pour faire pivoter l'image sans changer sa taille
def rotate_image(image, angle: float, fill_color=(255,255,255)):
    # Faire pivoter l'image
    rotated_image = image.rotate(angle, expand=False)
    # Créer une image de fond qui a la même taille que l'image originale
    background = Image.new(image.mode, rotated_image.size, fill_color)
    # Fusionner l'image de fond avec l'image pivotée
    background.paste(rotated_image, (0, 0), rotated_image)
    return background
```
