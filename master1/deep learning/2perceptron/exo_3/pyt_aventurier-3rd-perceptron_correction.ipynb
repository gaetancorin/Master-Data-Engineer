```python
import random
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Définir les données d'entraînement
compagnie_faible = [[8,8,8],[8,8,11],[11,11,8],[8,8,15],[11,11,11]]
compagnie_forte = [[11,15,8],[11,11,15],[15,15,8],[15,15,11],[15,15,15]]

valeurs_entree_cat0 = []
valeurs_entree_cat1 = []

for i in range(50) :
    # val = random.randint(0,4)
    for val in range(5) :
        valeurs_entree_cat0.append(compagnie_faible[val])
        valeurs_entree_cat1.append(compagnie_forte[val])

#print(len(valeurs_entree_cat0))
#print(len(valeurs_entree_cat1))

# Étiquettes correspondantes (0 pour la catégorie 0, 1 pour la catégorie 1)
etiquettes_cat0 = np.zeros(len(valeurs_entree_cat0), dtype = int)
etiquettes_cat1 = np.ones(len(valeurs_entree_cat1), dtype = int)

# Concaténer les données et les étiquettes
donnees_entree = np.concatenate([valeurs_entree_cat0, valeurs_entree_cat1])
print(donnees_entree)
etiquettes = np.concatenate([etiquettes_cat0, etiquettes_cat1])
print(etiquettes)

# Mélanger les données
indices_melanges = np.random.permutation(len(donnees_entree))
donnees_entree = donnees_entree[indices_melanges]
print(donnees_entree)
etiquettes = etiquettes[indices_melanges]
print(etiquettes)

# Créer le modèle
modele = Sequential()
modele.add(Dense(1, input_dim=3, activation='sigmoid'))

# Compiler le modèle
modele.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Entraîner le modèle
modele.fit(donnees_entree, etiquettes, epochs=400, batch_size=1)

# Évaluer le modèle
perte, precision = modele.evaluate(donnees_entree, etiquettes)
print(f"Perte : {perte}, Précision : {precision}")

# Prédictions
predictions = modele.predict(donnees_entree)
print("Prédictions:")
print(predictions)
```

```python
resultat = []
for i in range(10) :
    if predictions[i] < 0.5 :
        resultat.append('1 troll')
    else :
        resultat.append('2 troll')

print(predictions[0:9])
print(etiquettes[0:9])
print(resultat)
```
